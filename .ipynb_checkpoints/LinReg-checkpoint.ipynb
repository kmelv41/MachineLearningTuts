## Custom classes to be used in the Pipeline
### DummyEncoder
This comes from Tom Augspurgers PyData talk here: https://www.youtube.com/watch?v=KLPtEBokqQ0 
It's used to convert categorical variables to binary dummy variables.

### DataframeSelector
Takes the separeated categorical and numerical dataframes and converst them to separate numpy arrays.

### CombinedAttributesAdder
Adds a few more useful fields to data, including bedrooms per househould, population per household and bedrooms per room.

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import Imputer

class DummyEncoder(TransformerMixin):
    
    def fit(self, X, y=None):
        # record info here, use in transform, inv_transform.
        self.columns_ = X.columns
        self.cat_cols_ = X.select_dtypes(include=['category']).columns
        self.non_cat_cols_ = X.columns.drop(self.cat_cols_)
        
        self.cat_map_ = {col: X[col].cat for col in self.cat_cols_}
        left = len(self.non_cat_cols_) # 2
        self.cat_blocks_ = {}

        for col in self.cat_cols_:
            right = left + len(X[col].cat.categories)
            self.cat_blocks_[col] = slice(left, right)
            left = right
        return self
    
    def transform(self, X, y=None):
        return np.asarray(pd.get_dummies(X))
    
    def inverse_transform(self, trn, y=None):
        # Numpy to Pandas DataFrame
        # original column names <=> positions
        numeric = pd.DataFrame(trn[:, :len(self.non_cat_cols_)],
                               columns=self.non_cat_cols_)
        series = []
        for col, slice_ in self.cat_blocks_.items():
            codes = trn[:, slice_].argmax(1)
            cat = pd.Categorical.from_codes(codes,
                                            self.cat_map_[col].categories,
                                            ordered=self.cat_map_[col].ordered)
            series.append(pd.Series(cat, name=col))
        return pd.concat([numeric] + series, axis=1)[self.columns_]

# Create a class to select numerical or categorical columns 
# since Scikit-Learn doesn't handle DataFrames yet
# The DataFrameSelector class takes a list of attributes, turns it into a numpy array and drops the rest
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
    
    
# column index
rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

## Limit number of 'median_income' categories, split data

This cell turns the number of median_income categories from 15 into 5, since the data is skewed to the right.

It also splits the dataset into training and test splits using both random and stratified.

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit

# Divide by 1.5 to limit the number of income categories
df["income_cat"] = np.ceil(df["median_income"] / 1.5)
# Label those above 5 as 5
df["income_cat"].where(df["income_cat"] < 5, 5.0, inplace=True)

# This creates a completly random split of the data
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

# This code creates a stratified random set, stratified by median income
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df["income_cat"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
    
strat_train_set.dropna(inplace=True)
strat_test_set.dropna(inplace=True)
print(strat_train_set.head())
print(strat_test_set.head())
```

## Illustrate difference between stratified and random split

```python
# Check proportions of test set and training set
print(strat_test_set["income_cat"].value_counts() / len(strat_test_set))
print(strat_train_set["income_cat"].value_counts() / len(strat_train_set))

# Check the proportions of the pure random test set and training set
print(test_set["income_cat"].value_counts() / len(test_set))
print(train_set["income_cat"].value_counts() / len(train_set))

# Check the proportions of the total data set
df["income_cat"].value_counts() / len(df)
```

## Create the pipelines
Creates the two separate pipelines for categorical and numerical data, and a FeatureUnion pipeline to concatenate results.

```python
# Create two pipelines, one for numerical features and one for text features
# Pipelines run in parellel and are concatenated at the end of the pipeline

from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler
    
data_pipeline = Pipeline([
        ('dummy_encoder', DummyEncoder()),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

strat_train_prepared = data_pipeline.fit_transform(strat_train_set)
strat_train_prepared
```
